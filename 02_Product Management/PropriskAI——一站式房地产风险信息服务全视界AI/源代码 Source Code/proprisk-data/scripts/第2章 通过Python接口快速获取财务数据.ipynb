2.4 tushare获取数据案例实战

```python
#导入相关库
import tushare as ts
pro = ts.pro_api('75fe6c69bb10c996a556a402dde9fdf6e691dfdd2c30ff028ebd5a37')

# 需要获取数据的公司名称、股票代码、年份
comps = ['贵州茅台','五粮液','泸州老窖']
codes = ['600519.SH','000858.SZ','000568.SZ']
years = [2019,2018,2017]

# 循环获取语句
for i in range(len(comps)):
    for year in years:
        df_balance = pro.balancesheet(ts_code=codes[i],period=str(year)+'1231')
        df_income = pro.income(ts_code=codes[i],period=str(year)+'1231')
        df_cash = pro.cashflow(ts_code=codes[i],period=str(year)+'1231')
        # 保存为表格
        df_balance.to_excel('案例实战\\'+comps[i]+'_资产负债表_'+str(year)+'年.xlsx')  # to_excel函数将在下一章讲完pandas再详细讲解
        df_income.to_excel('案例实战\\'+comps[i]+'_利润表_'+str(year)+'年.xlsx')
        df_cash.to_excel('案例实战\\'+comps[i]+'_现金流量表_'+str(year)+'年.xlsx')
```

```python
# 转置显示表格，.T可以将DataFrame表格转置，关于DataFrame相关知识点请参考下一章
df_balance = df_balance.T
df_balance
```
